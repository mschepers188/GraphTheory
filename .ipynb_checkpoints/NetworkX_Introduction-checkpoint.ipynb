```python
nodes = ['A', 'B', 'C', 'D']
edges = [('A','B', {'weight':4}),
        ('A','C',{'weight':2}),
        ('B','C',{'weight':1}),
        ('B','D', {'weight':5}),
        ('C','D', {'weight':8})]

G = nx.Graph()
# for node in nodes:
#     G.add_node(node)
G.add_nodes_from(nodes)
G.add_edges_from(edges)
    
nx.draw_networkx(G, with_labels=True)
print(G.edges(data=True))
```
